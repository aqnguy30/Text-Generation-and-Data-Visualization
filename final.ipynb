```python
#Problem 1
f = open("mbox-short.txt", "r")
count = 1
for line in f:
    if "Received" in line:
        print(str(count) + " : " + line)
        count += 1
    if count > 10:
        break
```

```python
#Problem 2
import re
f = open("mbox-short.txt", "r")
count = 1
for line in f:
    date = re.findall(".{3} .{3}  [0-9]{1} [0-9]{2}:[0-9]{2}:[0-9]{2} [0-9]{4}", line)
    if date:
        date[0] = date[0].split()[0] + ", " + date[0].split()[2] + " " + date[0].split()[1] +  " " + date[0].split()[-1]
        print(count, date)
        count += 1
```

```python
#Problem 3
f = open("mbox-short.txt", "r")

out = []

for line in f:
    domain = re.findall("@[\w.]+", line)
    
    if domain and domain not in out and ("." in domain[0]):
        out.append(domain)

for count, item in enumerate(out , start=1):
    item[0] = item[0].strip("@")
    print(count, item)
```

```python
#import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#read-in the data
df = pd.read_excel("Superstore.xlsx")
df.head()
#np.shape(df)
```

```python
#Problem 4
df.describe()
```

```python
#Problem 5
q = pd.cut(df['Profit'],5)
grouped = df['Profit'].groupby(q)
def get_stats(group):
    return {'mean':group.mean(),'std':group.std(),'min':group.min(), 'max':group.max()}
grouped.apply(get_stats).unstack()
```

```python
#Problem 7
diff=lambda x: x.max()-x.min()
product_category_sale = df['Sales'].groupby(df['Product Category']).apply(diff)
product_category_sale.sort_values(ascending=False)
```
